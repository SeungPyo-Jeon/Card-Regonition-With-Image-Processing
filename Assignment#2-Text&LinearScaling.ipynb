```python
import cv2
import numpy as np

origin_rena = cv2.imread("./asset_imgs/rena.jpg",cv2.IMREAD_COLOR)

if origin_rena is None:
    raise Exception("Image File Read Error!")
    
target_rena = np.zeros( origin_rena.shape, origin_rena.dtype)
alpha = 1.5
beta = 50

for y in range( origin_rena.shape[0]):
    for x in range( origin_rena.shape[1]):
        for c in range( origin_rena.shape[2]):
            target_rena[y,x,c] = np.clip((alpha*origin_rena[ y, x, c]) + beta, 0, 255)
            
origin = "origin"
cv2.namedWindow(origin)
cv2.moveWindow(origin,0,0)
cv2.putText(origin_rena,"origin", (10,50), cv2.FONT_ITALIC, 3,(0,0,0),2)
cv2.imshow(origin,origin_rena)

target = "target"
cv2.namedWindow(target)
cv2.moveWindow(target,0,500)
cv2.putText( target_rena, "target",(10,50), cv2.FONT_ITALIC, 3, (0,0,0),2)
cv2.imshow(target, target_rena)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
origin_rena = cv2.cvtColor( origin_rena, cv2.COLOR_BGR2RGB)
target_rena = cv2.cvtColor( target_rena, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(origin_rena)
plt.title("origin_rena")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(target_rena)
plt.title("target_rena")
plt.axis('off')
plt.show()
```
