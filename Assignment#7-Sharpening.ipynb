```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

rena = cv2.imread("./asset_imgs/rena-s.jpg",cv2.IMREAD_GRAYSCALE)

if rena is None:
    print("Image Read Error")

padding_rena = np.zeros( (rena.shape[0]+2,rena.shape[0]+2), dtype= rena.dtype)
padding_rena[1:-1,1:-1] = rena

detail_mask = np.ones((3,3), dtype=np.uint8)*-1
detail_mask[1,1] = 9
detail_mask = detail_mask * (1/9)

detail = np.zeros( rena.shape, dtype= rena.dtype )
shaped_rena = np.zeros( rena.shape, dtype =rena.dtype )

print( rena.shape, padding_rena.shape, detail.shape )
#print( mask )

for y in range( 1,padding_rena.shape[0]-1):
    for x in range( 1,padding_rena.shape[1]-1):
        detail[y-1,x-1] = np.clip(np.sum(padding_rena[y-1:y+2,x-1:x+2] * detail_mask),0,255)
        
        
#for y in range()
cv2.imshow("rena",rena)
cv2.imshow("rena_detail",detail)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.subplot(121), plt.imshow(rena, cmap="gray"), plt.title('rena'), plt.axis('off')
plt.subplot(122), plt.imshow(detail, cmap="gray"), plt.title('rena_detail'), plt.axis('off')
plt.suptitle('sharpening')
plt.tight_layout()
plt.show()
```
