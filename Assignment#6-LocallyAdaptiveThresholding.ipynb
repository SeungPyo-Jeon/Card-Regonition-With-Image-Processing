```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
paper = cv2.imread("./asset_imgs/paper.jpg",cv2.IMREAD_GRAYSCALE)
if paper is None:
    print("Image Read Error")
    
window_size = [paper.shape[0]//10, paper.shape[1]//10]    
plt.figure( figsize=(24,30))
T = [ 10, 50,  100 ,125, 150, 175, 200 ]

LAT_paper = np.zeros( paper.shape, dtype = paper.dtype)
for dy in range( 10 ):
    for dx in range( 10 ):
        crop_paper = paper[ dy*window_size[0]: (dy+1)*window_size[0],
                          dx*window_size[1]: (dx+1)*window_size[1]].copy()
        ret, otsu_img = cv2.threshold(crop_paper, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
        
        LAT_paper[ dy*window_size[0]: (dy+1)*window_size[0],dx*window_size[1]: (dx+1)*window_size[1]] = otsu_img.copy()

#cv2.imshow("paper",paper)
#cv2.imshow("LAT",LAT_paper)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.subplot(121), plt.imshow(paper, cmap='gray'), plt.title('paper')
plt.axis('off')
plt.subplot(122), plt.imshow(LAT_paper, cmap='gray'), plt.title('LAT_paper')
plt.axis('off')
plt.tight_layout()
plt.suptitle('Locally Adaptive Thresholding', fontsize=20)
plt.show()
```
