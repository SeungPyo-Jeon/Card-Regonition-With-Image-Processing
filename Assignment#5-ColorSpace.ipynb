```python
import cv2 
import numpy as np
rena_bgr = cv2.imread("./asset_imgs/rena-s.jpg", cv2.IMREAD_COLOR)

if rena_bgr is None:
    print( "Image Read Error")
white = np.array( [255,255,255], np.uint8)

rena_cmy = white - rena_bgr
zero_img = np.zeros( (rena_cmy.shape[0],rena_cmy.shape[1]), dtype = np.uint8 )
rena_yellow = rena_cmy[:,:,0].astype(np.uint8).copy()
rena_magenta = rena_cmy[:,:,1].astype(np.uint8).copy()
rena_cyan = rena_cmy[:,:,2].astype(np.uint8).copy()

name = "BGR"
cv2.namedWindow(name)
cv2.putText(rena_bgr,"BGR",(10,30),cv2.FONT_ITALIC,1,(0,255,0),1)
cv2.imshow(name,rena_bgr)

name = "CMY"
cv2.namedWindow(name)
cv2.putText(rena_cmy,"CMY",(10,30),cv2.FONT_ITALIC,1,(0,255,0),1)
cv2.imshow("CMY",rena_cmy)

name = "Yellow"
cv2.namedWindow(name)
cv2.putText(rena_yellow,"Yellow",(10,30),cv2.FONT_ITALIC,1,(180.0),2)
cv2.imshow("Yellow",rena_yellow)

name = "magenta"
cv2.namedWindow(name)
cv2.putText(rena_magenta,"magenta",(10,30),cv2.FONT_ITALIC,1,(180.0),2)
cv2.imshow("magenta",rena_magenta)

name = "cyan"
cv2.namedWindow(name)
cv2.putText(rena_cyan,"cyan",(10,30),cv2.FONT_ITALIC,1,(180.0),2)
cv2.imshow("cyan",rena_cyan)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.subplot(231), plt.imshow(rena_bgr), plt.title('rena_bgr'), plt.axis('off')
plt.subplot(232), plt.imshow(rena_cmy), plt.title('rena_cmy'), plt.axis('off')
plt.subplot(234), plt.imshow(rena_yellow,cmap="gray"), plt.title('rena_yellow'), plt.axis('off')
plt.subplot(235), plt.imshow(rena_magenta,cmap="gray"), plt.title('rena_magenta'), plt.axis('off')
plt.subplot(236), plt.imshow(rena_cyan,cmap="gray"), plt.title('rena_cyan'), plt.axis('off')
plt.suptitle("Color Space Conversion")
plt.tight_layout()
plt.show()
```
