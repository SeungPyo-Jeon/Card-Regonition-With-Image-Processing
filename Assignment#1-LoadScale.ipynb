```python
#ReadImageFile
import cv2

gray2gray = cv2.imread("./asset_imgs/rena.jpg",cv2.IMREAD_GRAYSCALE)
gray2color = cv2.imread("./asset_imgs/rena.jpg",cv2.IMREAD_COLOR )

if gray2gray is None or gray2color is None:
    raise Exception("Image File Read Error!")

print("pixel information [100,100]")
print("%s %s" %("gray2gray", gray2gray[100,100]))
print("%s %s" %("gray2color", gray2color[100,100]))

cv2.imshow("gray2gray", gray2gray)
cv2.imshow("gray2color", gray2color)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
gray2color = cv2.cvtColor(gray2color, cv2.COLOR_BGR2RGB)
gray2gray = cv2.cvtColor(gray2gray, cv2.COLOR_GRAY2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow( gray2gray)
plt.title("gray2gray")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(gray2color)
plt.title("gray2color")
plt.axis('off')
plt.show()
```
